```python
# *************** NEURAL NETWORK MODEL **************************
```

```python
import sklearn.neural_network
```

```python
print(data.shape)
```

```python
X = data[["xc","yc","radius"]]
```

```python
Y = data.iloc[:,4:]
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
# scaler_test = StandardScaler()
```

```python
scaler.fit(X_train)
# scaler.fit(X_test)
```

```python
StandardScaler(copy=True, with_mean=True, with_std=True)
```

```python
# Now apply the transformations to the data:
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = sklearn.neural_network.MLPRegressor(hidden_layer_sizes = (5,),activation = 'logistic',
                    solver = 'adam', 
                    learning_rate = 'adaptive',
                    max_iter = 500,
                    learning_rate_init = 0.01,
                     alpha = 0.01)
```

```python
model.fit(X_train,Y_train)
```

```python
predictions = model.predict(X_test)
```

```python
print(predictions)
```

```python
# lOSS FUNCTION
sklearn.metrics.mean_squared_error(Y_test, predictions, sample_weight=None, multioutput="uniform_average")
```

```python
print(model.score(X_test,Y_test))
```

```python
model.predict(X_test[0:1])
```

```python
filename = 'finalized_model.sav'
scaler_file = 'scaler_param.sav'
```

```python
pickle.dump(model,open(filename,"wb"))
pickle.dump(scaler,open(scaler_file,"wb"))
```

```python
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, Y_test)
print(result)
```
